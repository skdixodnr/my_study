- Merge
  - key를 이용해 데이터의 row를 기준으로 연결시켜 합침(SQL의 join과 유사)

```python
import random
data1 = pd.DataFrame({'id':['01','02','03','04','05','06'], 'col1':np.random.randint(0,50,6),'col2':np.random.randint(1000,2000,6)})
data1
```

```python
data2 = pd.DataFrame({'id':['04','05','06','07'],'col1':np.random.randint(1000,5000,4)})
data2
```

```python
#inner join(교집합)
pd.merge(data1,data2,on='id') # on은 기준
```

```python
#how : 어떤 방식으로 병합할 것 인가?{inner,outer,left,right}
pd.merge(data1,data2,on='id',how='left')
```

```python
#outer join(합집합)
pd.merge(data1,data2,on='id',how='outer')
```

- cut

```python
ages = [20,53,17,7,23,64,6,2,45,91,21,56,12,3,4,6,26,52,51,41,23,75,91,5,14,23]
bins = [0,20,40,60,100]
```

```python
cuts = pd.cut(ages,bins)
cuts
```

```python
cuts.categories # cut 메소드의 결과는 Categorical이라는 특수 개체에 존재.
```

```python
cuts.codes
```

```python
# 구간을 균등한 길이로 나눔.
pd.cut(ages,4,precision=1).value_counts()
# precision : 소수점 자릿수, 3이면 소수점 세번쨰에서 반올림.
```

```python
# 구간을 균등한 길이로 나눔.
pd.qcut(ages,4).value_counts()
```

- 어떤 제작사가 제작한 영화의 총 수익이 가장 높은지를 알아보자.

- 원본 csv에서 달러 기호와 쉼표 기호와 심표 기호를 유지하기 위해 열의 값을 텍스트로 가져옴.
- 열의 값을 십진수로 변환하려면 두 가지 기호('$',',')를 모두 제거하고 빈 텍스트로 바꾸어야 함.

```python
df['Gross'].str.replace("$","",regex = False).str.replace(",","", regex = False)
```

- 참고 : 정규표현식(https://hamait.tistory.com/342)

```python
df['Gross'] = (
      df['Gross']
      .str.replace("$","",regex = False)
      .str.replace(",","", regex = False)
      .astype(float)
)
```

```python
df['Gross'].mean()
```

- Face color

```python
x = np.arange(10)
y1 = x
y2 = x**2 -x

fig,axs = plt.subplots(2,1) # 2행 1열로 그리겠습니다.
fig.set_facecolor('#bd28bf') # 뒤의 배경

axs[0].plot(x,y1) # 첫번째 행
axs[1].plot(x,y2) # 두번째 행

axs[0].set_facecolor('pink') # 첫번째 행의 배경
axs[1].set_facecolor('skyblue') # 두번째 행의 배경
```
